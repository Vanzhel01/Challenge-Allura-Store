```python

heat_data = tienda_coord[['lat', 'lon']].values.tolist()
map_center = [tienda_coord['lat'].mean(), tienda_coord['lon'].mean()]

m = fl.Map(
    location=map_center,
    zoom_start=6,
    tiles="CartoDB positron"
)
HeatMap(heat_data, radius=9).add_to(m)
```
